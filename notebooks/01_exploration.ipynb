### Nous allons essayer de voir si les patients hypertendus ont plus fait l'AVC et de même pour les patients atteints de maladie cardiaque via une analyse exploratoire pour étudier la relation entre deux variables catégorielles.

```python
# AVC selon hypertension
pd.crosstab(df["hypertension"], df["stroke"], normalize='index').round(3)
```

En comparant le taux d'AVC nous avons :

Non hypertendus : 4,0 %
Hypertendus : 13,3 %
Les patients hypertendus présentent donc un risque d'AVC.

```python
# la différence entre les 2 taux
print(f"Les patients hypertendus présentent donc un risque d'AVC : {13.3/4} fois plus élévés") 
```

Les patients souffrant d'hypertension présentent une proportion d'AVC nettement supérieure à celle observée chez les patients non hypertendus (13,3 % contre 4,0 %). 
L'hypertension apparaît ainsi comme un facteur fortement associé à la survenue d'un AVC.

```python
# AVC selon maladie Cradiaque 
pd.crosstab(df["heart_disease"], df["stroke"], normalize='index').round(3)
```

En comparant le taux d'AVC :

Sans maladie cardiaque : 4,2 %
Avec maladie cardiaque : 17,0 %
Les patients atteints d'une maladie cardiaque présentent donc un risque d'AVC.

```python
# la différence entre les 2 taux
print(f"Les patients atteints d'une maladie cardiaque présentent donc un risque d'AVC : {(17.0/4.2)} fois plus élévés") 
```

Les patients atteints d'une maladie cardiaque présentent un taux d'AVC d'environ 17 %, contre seulement 4,2 % chez les patients sans antécédent cardiaque. 
La présence d'une maladie cardiaque semble donc fortement associée au risque d'AVC.

```python
# glycémie par AVC
df.groupby("stroke")["avg_glucose_level"].mean().round(3)
```

Les patients ayant subi un AVC présentent une glycémie moyenne significativement plus élevée : 132,5 contre 104,8.
La glycémie apparaît comme un facteur potentiellement associé au risque d'AVC.

```python
# IMC par AVC
df.groupby("stroke")["bmi"].mean().round(3)
```

L'IMC moyen chez les patients ayant subi un AVC est légèrement supérieur à celui des autres patients. 30,5 contre 28,8.
Une association semble exister mais apparaît moins marquée.


```python
# Tabagisme par AVC
pd.crosstab(df["smoking_status"], df["stroke"], normalize='index').round(3)
```

Le taux le plus élevé est observé chez Formerly smoked avec 7,9 % contre 4.8 % chea les non-fumeurs.
On observe une association entre fumer et AVC mais il y'a certainement d'autres facteurs liés pour parler de causalité.
